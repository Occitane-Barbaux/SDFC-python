## Testing on full algorithm
Only with GEV
### MH

```python
import SDFC as sd
```

```python
import numpy as np
import scipy.stats as sc
```

```python


n_samples=1000
n_mcmc_drawn=1000
prior_law=sc.multivariate_normal( mean = [2,1,-0.1] , cov = np.identity(3)*(0.1) , allow_singular = True )
```

```python


kwargs = { "n_samples" : n_samples ,
"name"    : "GEV" ,
"sd_law"  : sd.GEV() ,
"params"  : ["loc","scale","shape"] ,
"shape_p" : False}

```

```python
#Ok run. Fixed, one scale
kwargs["transition_type"]="Fixed"
kwargs["fixed_cov"]=np.array([0.5])
ns_law=sd.GEV()
Yo     = sc.genextreme( loc = 2 ,  scale =0.5 , c = - 0.1 ).rvs(150)
ns_law._method="bayesian"
ns_law.fit( Yo , n_mcmc_drawn =  n_mcmc_drawn , prior=prior_law , **kwargs )
ns_law.loc
```

```python
#Ok run. Fixed, dif scale
kwargs["transition_type"]="Fixed"
kwargs["fixed_cov"]=np.array([0.5,0.1,0.2])
ns_law=sd.GEV()
Yo     = sc.genextreme( loc = 2 ,  scale =0.5 , c = - 0.1 ).rvs(150)
ns_law._method="bayesian"
ns_law.fit( Yo , n_mcmc_drawn =  n_mcmc_drawn , prior=prior_law , **kwargs )
ns_law.loc
```

```python
#Ok run. adaptative
kwargs["transition_type"]="Adapt"

ns_law=sd.GEV()
Yo     = sc.genextreme( loc = 2 ,  scale =0.5 , c = - 0.1 ).rvs(150)
ns_law._method="bayesian"
ns_law.fit( Yo , n_mcmc_drawn =  n_mcmc_drawn , prior=prior_law , **kwargs )
ns_law.loc
```

# ESS stopping point
No fixed length of the algorithm, stop when enough samples

```python
import SDFC as sd
import numpy as np
import scipy.stats as sc
```

```python
n_samples=1000
n_mcmc_drawn=10000
prior_law=sc.multivariate_normal( mean = [2,1,-0.1] , cov = np.identity(3)*(0.1) , allow_singular = True )
kwargs = { "n_samples" : n_samples ,
"name"    : "GEV" ,
"sd_law"  : sd.GEV() ,
"params"  : ["loc","scale","shape"] ,
"shape_p" : False,
"n_ess"   : 10,
"burn_in" : 800        }
```

```python
#Ok run. Fixed, all scales 
kwargs["transition_type"]="Fixed"
kwargs["fixed_cov"]=np.array([0.5,0.2,0.3])
ns_law=sd.GEV()
Yo     = sc.genextreme( loc = 2 ,  scale =0.5 , c = - 0.1 ).rvs(150)
ns_law._method="bayesian-experimental-ESS"
ns_law.fit( Yo  , prior=prior_law , **kwargs )
ns_law.loc
```

```python
#Ok run. adaptative
kwargs["transition_type"]="Adapt"

ns_law=sd.GEV()
Yo         = sc.genextreme( loc = 2 ,  scale =0.5 , c = - 0.1 ).rvs(150)
ns_law._method="bayesian-experimental-ESS"
ns_law.fit( Yo , prior=prior_law , **kwargs )
ns_law.loc
```

```python
sd.Para_Runs(ns_law,prior_law,MLE_theta=["","",""],pathOut="",q_l=0.05,q_h=0.95,show=True)
sd.Summary_run_table(ns_law,prior_law,MLE_theta=["","",""],pathOut="",q_l=0.05,q_h=0.95,show=True)
```
